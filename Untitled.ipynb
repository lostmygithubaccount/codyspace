```python
import os
import requests
```

```python
headers = {"Authorization": f"token {os.environ.get('CODESPACE_PAT')}"}
```

```python
url = "https://api.github.com/orgs/dbt-labs/teams/core/members"
r = requests.get(url, headers=headers)
rj = r.json()

exclude_logins = [login["login"] for login in rj]
print(exclude_logins)
```

```python
# graphql query to get list of issues for the github repo
query = """
    query {
        repository(owner:"dbt-labs", name:"dbt-core") {
            issues(last:100 states:OPEN) {
                edges {
                    node {
                        title
                        assignees(first:10) {
                            edges {
                                node {
                                    login
                                }
                            }
                        }
                        number
                        url
                        labels(first:10) {
                            edges {
                                node {
                                    name
                                }
                            }
                        }
                        participants(first:100) {
                            edges {
                                node {
                                    login
                                }
                            }
                        }
                        createdAt
                        updatedAt
                        closedAt
                        author {
                            login
                        }
                    }
                }
            }
        }
    }
"""

url = "https://api.github.com/graphql"
json = {"query": query}

r = requests.post(url, json=json, headers=headers)
```

```python
rj = r.json()
issues = rj["data"]["repository"]["issues"]["edges"]
```
