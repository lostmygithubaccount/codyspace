```python
import os
import requests
```

```python
headers = {"Authorization": f"token {os.environ.get('CODESPACE_PAT')}"}
```

```python
ORG = "dbt-labs"
REPOS = [
    #"dbt-core",
    "dbt-snowflake",
    "dbt-redshift",
    "dbt-bigquery",
    "dbt-rpc",
    "dbt-server",
    "dbt-spark",
    "dbt-starter-project",
    "dbt-utils",
    "homebrew-dbt",
]
```

```python
url = f"https://api.github.com/repos/{ORG}/dbt-core/labels"

core_labels = []

for page in range(10):
    try:
        params = {"per_page": 100, "page": page}

        r = requests.get(url, headers=headers, params=params)
        rj = r.json()

        core_labels.extend(rj)
    except:
        break

core_labels[0]
```

```python
len(core_labels)
```

```python
# I'm a baller
core_label_color_map = {label['name']: label['color'] for label in core_labels}

core_label_color_map
```

```python
url_base = "https://api.github.com/repos"

for repo in REPOS:
    print(repo)
    for label in core_label_color_map:
        url = f"{url_base}/{ORG}/{repo}/labels/{label}"
        r = requests.get(url, headers=headers)
        if r.status_code == 200:
            print(f"{label} exists")

            lj = r.json()

            eq_core = lj['color'] == core_label_color_map[label]

            print(f"Equal: {eq_core} | {repo}: {lj['color']} | dbt-core: {core_label_color_map[label]}")

            if not eq_core:
                data = {
                    "color": core_label_color_map[label]
                }

                r = requests.patch(url, headers=headers, json=data)
                print(r.status_code)
                print(r.text)
```
